```
import numpy as np
import matplotlib.pyplot as plt
```

```
data_norm_mod = np.load('/content/drive/MyDrive/att_10617_models/history_mod.npz')
data_att_mod = np.load('/content/drive/MyDrive/att_10617_models/history_att_mod.npz')
#############################################
data_norm = np.load('/content/drive/MyDrive/norm_10617/history.npz')
data_att = np.load('/content/drive/MyDrive/project_models_10617/history_att.npz')
```

```
train_loss_norm_mod = data_norm_mod['train_loss']
val_loss_norm_mod = data_norm_mod['val_loss']

train_loss_att_mod = data_att_mod['train_loss']
val_loss_att_mod = data_att_mod['val_loss']

#############################################
train_loss_norm = data_norm['train_loss']
val_loss_norm = data_norm['val_loss']

train_loss_att = data_att['train_loss']
val_loss_att = data_att['val_loss']

#############################################
n_units = data_norm_mod['n_units']
```

# Normal Modded Model


```
plt.plot(train_loss_norm_mod[0], label='train_loss_mod_{}'.format(n_units[0]))
plt.plot(train_loss_norm_mod[1], label='train_loss_mod_{}'.format(n_units[1]))

plt.plot(val_loss_norm_mod[0], label='val_loss_mod_{}'.format(n_units[0]))
plt.plot(val_loss_norm_mod[1], label='val_loss_mod_{}'.format(n_units[1]))

plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve - Modded Normal Model')
plt.show()

print(n_units[0], min(val_loss_norm_mod[0]),np.argmin(val_loss_norm_mod[0])+1, n_units[1], min(val_loss_norm_mod[1]), np.argmin(val_loss_norm_mod[1])+1)
```

# Attention Modded Model



```
plt.plot(train_loss_att_mod[0], label='train_loss_att_mod_{}'.format(n_units[0]))
plt.plot(train_loss_att_mod[1], label='train_loss_att_mod_{}'.format(n_units[1]))

plt.plot(val_loss_att_mod[0], label='val_loss_att_mod_{}'.format(n_units[0]))
plt.plot(val_loss_att_mod[1], label='val_loss_att_mod_{}'.format(n_units[1]))

plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve - Modded Attention Model')
plt.show()

print(n_units[0], min(val_loss_att_mod[0]), np.argmin(val_loss_att_mod[0])+1, n_units[1], min(val_loss_att_mod[1]), np.argmin(val_loss_att_mod[1])+1)
```
